```python
# Importando as bibliotecas
import pandas as pd
import numpy as np
from matplotlib import pyplot as plt
import seaborn as sns
from matplotlib.patches import Patch


# Importando o data frame
df = pd.read_csv('https://drive.google.com/uc?export=download&id=1DloAUKJym4ctB5AgS0g2m4rqwUWukI0Y')
df
```

```python
# Transformar target em 1 ou 0
replacer = { 'y' : 1, 'yes': 1, 'n' : 0 , 'no' : 0}
df[['y']] = df[['y']].replace(replacer)

# Transformar features binárias em Sim ou Não
replacer_2 = { 'y' : "Sim", 'yes': "Sim", 'n' : "Não" , 'no' : "Não"}
df[['default','housing','loan']] = df[['default','housing','loan']].replace(replacer_2)

# Separando as colunas em categóricas e numéricas
dados_numericos = ['id','age','balance','day','duration','campaign','pdays','previous', 'y']
dados_categoricos = ['job','marital','education','location','contact','month','poutcome', 'y']
df_num = df[dados_numericos]
df_cat = df[dados_categoricos]
```

# Visualizar distribição de features categóricas


```python
# Barchart - Emprego
sns.countplot(data = df, y = "job", palette=sns.color_palette("Spectral", 13)
, legend=False)
plt.title("Distribuição - Emprego")
plt.show()

# Barchart - Localização
sns.countplot(data = df, y = "location", palette="Spectral")
plt.title("Distribuição - Localização")
plt.show()

# Barchart - Poutcome
sns.countplot(data = df, y = "poutcome", palette="Spectral")
plt.title("Distribuição -  Resultado da campanha de marketing anterior")
plt.show()

# Barchart - Contact
sns.countplot(data = df, y = "contact", palette="Spectral")
plt.title("Distribuição - Tipo de comunicação")
plt.show()

# Barchart - Default
sns.countplot(data = df, y = "default", palette="Spectral")
plt.title("Distribuição - Cliente possui inadimplência")
plt.show()

# Barchart - loan
sns.countplot(data = df, y = "loan", palette="Spectral")
plt.title("Distribuição - Cliente  possui um empréstimo ativo")
plt.show()

# Barchart - Housing
sns.countplot(data = df, y = "housing", palette="Spectral")
plt.title("Distribuição - Cliente  possui empréstimo habitacional")
plt.show()

# Barchart - Estado civil
sns.countplot(data = df, x = "marital", palette="Spectral")
plt.title("Distribuição - Estado civil")
plt.show()

# Barchart - Month
sns.countplot(data = df, x = "month", palette="Spectral")
plt.title("Distribuição -  Mês do último contato")
plt.show()

# Barchart - Educação
sns.countplot(data = df, x = "education", palette="Spectral")
plt.title("Distribuição -  Educação")
plt.show()

```

# Relações entre features categóricas e target



```python
def plot_countplot(df, x_col, title):
    sns.countplot(data=df, x=x_col, hue='y', palette={0: 'red', 1: 'green'})
    plt.title(title)
    handles, labels = plt.gca().get_legend_handles_labels()
    new_labels = ['Sucesso da Campanha' if label == '1' else 'Fracasso da Campanha' for label in labels]
    plt.legend(handles, new_labels, title='Resultado')
    plt.xticks(rotation=45)
    plt.show()
plots_info = [
    ('job', 'Distribuição do trabalho em função do resultado da campanha'),
    ('marital', 'Distribuição do estado civil em função do resultado da campanha'),
    ('education', 'Distribuição da educação em função do resultado da campanha'),
    ('location', 'Distribuição da localização em função do resultado da campanha'),
    ('contact', 'Distribuição do tipo de comunicação em função do resultado da campanha'),
    ('month', 'Distribuição do mês do último contato em função do resultado da campanha'),
    ('poutcome', 'Distribuição de campanha anterior em função do resultado da campanha'),
    ("default", "Distribuição - Cliente possui inadimplência"),
    ("loan", "Distribuição - Cliente  possui um empréstimo ativo"),
    ("housing", "Distribuição - Cliente  possui empréstimo habitacional")
]
for x_col, title in plots_info:
    plot_countplot(df, x_col, title)
```

# Distribuição das features numéricas


```python
# Histograma com Linha KDE
# Age
sns.histplot(data=df, x='age', binwidth=5, kde=True)
plt.title("Idade")
plt.xlabel('Idade')
plt.show()

# Balance
sns.histplot(data=df, x='balance', kde=True)
plt.title("Balanço médio anual, em euros")
plt.xlabel('Balanço médio anual')
plt.show()

# Duration
sns.histplot(data=df, x='duration', binwidth=10, kde=True)
plt.title("Duração do último contato, em minutos")
plt.xlabel('Duração')
plt.show()

# Campaign
sns.histplot(data=df, x='campaign', binwidth=1, kde=True)
plt.title("Número de contatos realizados com o cliente durante a campanha")
plt.xlabel('Número de contatos')
plt.show()

# Day
sns.histplot(data=df, x='day', binwidth=1, kde=True)
plt.title("Dia do último contato")
plt.xlabel('Dia')
plt.show()

# Pday
df_filtrado = df_num[df['pdays'] != -1]
sns.histplot(x=df_filtrado["pdays"], binwidth=20, kde=True)
plt.title(" Número de dias passados desde o contato em campanha anterior,")
plt.xlabel('Dias passados')
plt.show()
```

# Relações entre features numéricas e target

```python
paleta = {0: 'red', 1: 'green'}

g = sns.PairGrid(df_num, hue="y", palette=paleta)
g.map_diag(sns.histplot)
g.map_offdiag(sns.scatterplot, palette=paleta)

handles = [
    Patch(color='red', label='Fracasso na campanha'),
    Patch(color='green', label='Sucesso na campanha')
]
g.add_legend(title='Campanha', handles=handles)

plt.show()
```

```python
# Heatmap com features numéricas
sns.heatmap(df_num.corr(), annot=True, fmt=".2f")
plt.title("Matriz de correlações para features do dataset")
plt.show()
```

```python
# Dicionário com os títulos específicos para cada feature
titles = {
    'age': 'Distribuição Estatística da Idade em função do resultado da campanha',
    'balance': 'Distribuição Estatística do balanço médio anual (em euros) em função do resultado da campanha',
    'day': 'Distribuição Estatística do dia do último contato em função do resultado da campanha',
    'duration': 'Distribuição Estatística da Duração do último contato (em minutos) em função do resultado da campanha',
    'campaign': 'Distribuição Estatística do número de contatos realizados durante a campanha em função do resultado da campanha',
    'pdays': 'Distribuição Estatística dos número de dias passados desde contato em campanha anterior em função do resultado da campanha',
    'previous': 'Distribuição Estatística do número de contatos anteriores em função do resultado da campanha'
}

features = ['age', 'balance', 'day', 'duration', 'campaign', 'pdays', 'previous']

for feature in features:
    plt.figure(figsize=(8, 6))
    sns.boxplot(data=df, y=feature, x="y", hue="y", palette={0: 'red', 1: 'green'})

    # Legenda
    legend_patches = [
        Patch(color='red', label='Fracasso da campanha'),
        Patch(color='green', label='Sucesso da campanha')
    ]
    plt.legend(handles=legend_patches, title='Resultado da campanha', loc='upper right')

    # Definir o título do gráfico
    plt.title(titles.get(feature, 'Título padrão'))

    plt.show()
```
